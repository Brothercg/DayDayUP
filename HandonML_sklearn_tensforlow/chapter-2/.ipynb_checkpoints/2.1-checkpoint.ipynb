```python
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

+ 其他的许多数据都会存在上限的情况。包括数据的标签，例如此处使用的房价，如果应用这些数据进行训练，得出的模型会认为房价会有一个上限$value$，例如训练数据中的最大值。使用这些模型的时候，必须要考虑这是否会成为一个问题，如果需要预测一个房价远远超过$value$的问题，有两个处理方式：
    + 收集这些极限数据的合适的特征。
    + 从训练集中删除这些数据，与此同时删除测试集中的数据，这样，模型便不再能处理这种特殊的数据。
+ 这些属性的比重也是不一样的。这点将在随后讲到。
+ 同时上述许多直方图，具有长尾效应，即在数据的中位数的右侧，分布的比左侧要长一些，即中位数左侧的种类要少于右侧。对于某些机器学习算法，想要学习出这样的特征是比较困难的。


## 设定随机数种子，自己写划分测试集与训练集

 构造测试集

```python
import numpy as np

def split_train_test(data, test_radio, seed=42):
    
    #添加随机种子，这样本方法在运行的时候就不会出现产生不同测试集与训练集的情况。
    np.random.seed(seed)
    
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_radio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, test_radio=0.2)
```

```python
# 添加了随机种子之后，运行过一次的代码的测试集与训练集的划分就不会因为在此运行而发生改变。
train_set.head()
```

## 给每一条数据绑定一个id来划分测试集与训练集
但是上述方法在数据集发生更新之后也许会发生问题。方法是给每一个测试样例安装一个id。然后基于id来划分测试集与训练集

```python
from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio *2**32

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

如果在数据不够多的情况下，便需要分层抽样：将人口划分为称为分层的同质子群，并从每个层中抽取正确数量的实例，以保证测试集代表总人口，如果随机采样的话，便会有采样不均匀的情况。

对于不同层级的数据，拥有充足的样本是十分重要的。

# 查看数据的内在关系
标准相关系数，又名皮尔森相关系数：
$$
\rho_{XY} = \frac{COV(X,Y)}{\sigma_X \sigma_Y} 
$$ 

+ X,Y之间皮尔森先关系数的值越接近1，表明两者越是接近正相关的关系，例如收入增加的话，房价也跟着增加。
+ 相反，越接近-1，X,Y之间便越接近负相关。

皮尔森相关系数只能衡量线性相关性，对于衡量非线性模型，其完全失效。

# 用户自定义数据处理

```python
housing.head()
```

```python
housing.values
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, householeds_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
        
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, householeds_ix]
        population_per_household = X[:, population_ix] / X[:, householeds_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

# 转换通道
到此为止，会发现有许多的数据需要进行正确的数据转换流程。

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipline = Pipeline([
    ("imputer" , SimpleImputer(strategy="median")),    # 处理数据缺失,并用对应的缺失属性的中位数代替。
    ('attribs_adder', CombinedAttributesAdder()),      # 添加用户自定义的数据处理
    ('std_scaler', StandardScaler()),                  # 表幻化
])

housing_num_tr = num_pipline.fit_transform(housing_num)
```

```python
type(housing_num_tr)
```

```python
housing_num_tr[0]
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

# 这里只是数据的格式吧？
full_pipeline = ColumnTransformer([
    ("num", num_pipline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])

#这里是传入数据，并对数据按照格式进行处理。
housing_prepared = full_pipeline.fit_transform(housing)
```

```python
from scipy import stats

confidence = 0.95
squared_errors = (final_predictions - y_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1,
                         loc=squared_errors.mean(),
                         scale=stats.sem(squared_errors)))
```

```python
m = len(squared_errors)
mean = squared_errors.mean()
tscore = stats.t.ppf((1 + confidence) / 2, df=m - 1)
tmargin = tscore * squared_errors.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - tmargin), np.sqrt(mean + tmargin)
```

```python
zscore = stats.norm.ppf((1 + confidence) / 2)
zmargin = zscore * squared_errors.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - zmargin), np.sqrt(mean + zmargin)
```
